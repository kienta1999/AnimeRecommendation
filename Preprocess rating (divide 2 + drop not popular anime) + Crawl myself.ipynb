## Crawl kienta1999

```python
from jikanpy import Jikan
jikan = Jikan()

kien = jikan.user(username='kienta1999', request='animelist')
my_id = 9147556
```

```python
kien_ratings = pd.DataFrame(columns = ['user_id', 'movie_id', 'rating'])
for i in kien['anime']:
    entry = {'user_id': 9147556, 'movie_id': i['mal_id'], 'rating': int(i['score']) / 2}
    #print(str(i['mal_id']) + ' ' + str(i['title']) + ' ' + str(i['score']))
    kien_ratings = kien_ratings.append(entry,ignore_index=True)
pd.set_option('display.max_columns', 200)
kien_ratings = kien_ratings[kien_ratings['rating'] != 0]
ratings = ratings.append(kien_ratings)
ratings.reset_index(drop = True, inplace = True)
#type(ratings)
```

```python
ratings[['user_id', 'movie_id']] = ratings[['user_id', 'movie_id']].astype(int)
ratings
```

```python
ratings.to_csv('my_user_ratings_full_3m.csv', index = False)
```
