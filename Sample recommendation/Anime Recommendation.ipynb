```python
import pandas as pd
from surprise import SVD
from surprise import accuracy
from surprise import Reader
from surprise import Dataset
from surprise.model_selection import train_test_split
```

```python
reader = Reader(line_format='user item rating', sep=',', skip_lines=1)
data = Dataset.load_from_file('my_user_ratings_divide2_dropunpopular.csv', reader=reader)
```

```python
algo = SVD(n_epochs = 30, lr_all = 0.005, n_factors = 50)
train, test = train_test_split(data, test_size=.0000000001)
```

```python
algo.fit(train)
```

```python
popular_anime = pd.read_csv('All anime with 40000 ratings.csv')
#popular_anime = popular_anime[popular_anime['ratingCount'] > 200000]
```

```python
kien = pd.DataFrame(columns = ['movieId', 'title', 'ratings'])
kien['movieId'] = popular_anime['id'].copy()
kien['title'] = popular_anime['title'].copy()
for i in kien.index:
    movie_id = kien.loc[i, 'movieId']
    rating = algo.predict('9147556', str(movie_id) , verbose=True)[3] * 2
    kien.loc[i, 'ratings'] = rating
```

```python
kien.sort_values(by = 'ratings', inplace = True)
```
